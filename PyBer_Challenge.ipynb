### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"])["ride_id"].count()
total_rides_title = pd.Series(total_rides, name = "Total Rides")
total_rides_title
```

```python
# 2. Get the total drivers for each city type NOT CORRECT YET
total_drivers = city_data_df.groupby(["type"])["driver_count"].sum()
total_drivers_title = pd.Series(total_drivers, name = "Total Drivers")
total_drivers_title
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"])["fare"].sum()
total_fares_title = pd.Series(total_fares, name = "Total Fares")
total_fares_title
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_ride = total_fares / total_rides
average_fare_ride = total_fares_title / total_rides_title
average_fare_ride_title = pd.Series(average_fare_ride, name = "Average Fare per Ride")
average_fare_ride_title
```

```python
# 5. Get the average fare per driver for each city type. NOT CORRECT YET
average_fare_driver = total_fares / total_drivers
average_fare_driver = total_fares_title / total_drivers_title
average_fare_driver_title = pd.Series(average_fare_driver, name = "Average Fare per Driver")
average_fare_driver_title
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.concat([total_rides_title, total_drivers_title, total_fares_title, average_fare_ride_title, average_fare_driver_title], axis = 1)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${0:,.2f}".format)

pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${0:,.2f}".format)

pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${0:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame

%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

fares_sum_df = pyber_data_df.groupby(["type", "date"])[["fare"]].sum()
fares_sum_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_sum_df = fares_sum_df.reset_index()
fares_sum_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fares_sum_df = fares_sum_df.pivot(index = "date", columns = "type", values = "fare")
fares_sum_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

jan_apr_df = fares_sum_df.loc["2019-01-01":"2019-04-29"]
jan_apr_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
jan_apr_df.index = pd.to_datetime(jan_apr_df.index)
jan_apr_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
jan_apr_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = jan_apr_df.resample("W").sum()
weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
ax = weekly_fares_df.plot(figsize=(14, 4))
ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare ($USD)")
ax.legend(loc = "center")
ax.plot(weekly_fares_df)
plt.savefig("analysis/Total_Fare_by_Type.png")

```
